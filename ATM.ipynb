```python
import random

class Atm:
    def __init__(self):
        self.User_name = "AMAN MODI"
        self._Balance = 10000.00  # Starting balance for the user
        self.old_pin = '12345'
        self.transaction_history = []  # List to store transaction history
        self.failed_attempts = 0
        self.locked = False
        print("Welcome", self.User_name)
        self.login()

    def login(self):
        """Function to handle user login"""
        if self.locked:
            print("Account is locked due to multiple failed login attempts.")
            return

        for attempt in range(3):
            pin = input("\nEnter Your PIN: ")
            if self.old_pin == pin:
                print("Login successful!")
                self.menu()
                break
            else:
                self.failed_attempts += 1
                print(f"Incorrect PIN. {3 - self.failed_attempts} attempts remaining.")
                if self.failed_attempts >= 3:
                    print("Account is locked due to multiple failed login attempts.")
                    self.locked = True
                    break

    def menu(self):
        """ATM menu with additional options"""
        if self.locked:
            return

        print('''
        1. Press 1 for Check balance
        2. Press 2 for Withdrawal
        3. Press 3 for Pin Generation
        4. Press 4 for Deposit
        5. Press 5 for View Transaction History
        6. Press 6 for Transfer Money
        7. Press 7 for Change PIN
        8. Press 8 to Exit
        ''')

        try:
            choice = int(input("Enter Your choice: "))
            if choice == 1:
                self.check_bal()
            elif choice == 2:
                self.withdrawal()
            elif choice == 3:
                self.pin_gen()
            elif choice == 4:
                self.deposit()
            elif choice == 5:
                self.view_transactions()
            elif choice == 6:
                self.transfer_money()
            elif choice == 7:
                self.change_pin()
            elif choice == 8:
                print("Thank you for using the ATM. Goodbye!")
                exit()
            else:
                print("Invalid choice. Please try again.")
                self.menu()
        except ValueError:
            print("Invalid input. Please enter a number between 1 and 8.")
            self.menu()

    def check_bal(self):
        """Function to check account balance"""
        print(f"Your account balance is: {self._Balance} Cr")
        self.menu()

    def withdrawal(self):
        """Function to handle withdrawal"""
        try:
            amount = float(input("Enter withdrawal amount: "))
            if amount > self._Balance:
                print("Insufficient balance.")
            else:
                otp = random.randint(100000, 999999)
                print(f"Your OTP for withdrawal is: {otp}")
                user_otp = int(input("Enter the OTP: "))
                if user_otp == otp:
                    self._Balance -= amount
                    self.transaction_history.append(f"Withdrawal: {amount} Cr")
                    print(f"Withdrawal successful. Your new balance is: {self._Balance} Cr")
                else:
                    print("Invalid OTP.")
        except ValueError:
            print("Invalid amount entered. Please try again.")
        self.menu()

    def pin_gen(self):
        """Function to generate a new PIN"""
        for i in range(3):
            old_pin = input("\nEnter Your Old Pin: ")
            if self.old_pin == old_pin:
                new_pin = input("Enter Your New Pin: ")
                self.old_pin = new_pin
                print("PIN successfully changed.")
                break
            else:
                print("Incorrect old PIN. Please try again.")
        self.menu()

    def deposit(self):
        """Function to deposit money"""
        try:
            credit = float(input("Enter your deposit amount: "))
            self._Balance += credit
            self.transaction_history.append(f"Deposit: {credit} Cr")
            print(f"Deposit successful. Your new balance is: {self._Balance} Cr")
        except ValueError:
            print("Invalid amount entered. Please try again.")
        self.menu()

    def view_transactions(self):
        """Function to view transaction history"""
        if self.transaction_history:
            print("\nTransaction History:")
            for transaction in self.transaction_history:
                print(transaction)
        else:
            print("No transactions made yet.")
        self.menu()

    def transfer_money(self):
        """Function to transfer money between accounts"""
        try:
            recipient = input("Enter recipient's name: ")
            amount = float(input("Enter amount to transfer: "))
            if amount > self._Balance:
                print("Insufficient balance to make the transfer.")
            else:
                self._Balance -= amount
                self.transaction_history.append(f"Transferred {amount} Cr to {recipient}")
                print(f"Transfer successful. Your new balance is: {self._Balance} Cr")
        except ValueError:
            print("Invalid amount entered. Please try again.")
        self.menu()

    def change_pin(self):
        """Function to change PIN"""
        new_pin = input("Enter your new PIN: ")
        self.old_pin = new_pin
        print("Your PIN has been successfully changed.")
        self.menu()


# To start the ATM, create an object of the Atm class:
atm = Atm()
```
